```python
import pandas as pd
import numpy as np 
df= pd.DataFrame({'From_To':['LoNDon_paris','MAdrid_miLAN','londON StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
                 'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                 'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                 'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
df
```

# Some values in the the FlightNumber column are missing. 
These numbers are meant to increase by 10 with each row so 10055 and 10075 need to be put in
place. Fill in these missing numbers and make the column an integer column
(instead of a float column).

```python
#Fill  these missing numbers by increasing 10 with preceding value
for i in np.arange(0,df.From_To.count()):
    
    if(pd.isnull(df['FlightNumber'].loc[i,])):
        df.loc[i,'FlightNumber']=df['FlightNumber'].loc[i-1] + 10
df['FlightNumber']
df
    
```

```python
#covert the float datatype to int 
df['FlightNumber'].astype(int)
```
